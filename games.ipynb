```python
import pandas as pd
import altair as alt

data = pd.read_csv("./Databases/compiled_db.csv")
data.head()
```

```python
# Implementing Overview+Detail using semantic zoom
selection = alt.selection(type="multi", fields=["Year Released"])
base = alt.Chart(data).properties(width=500, height = 250)

#Specify our overview chart
overview = alt.Chart(data).mark_bar().encode(
    y = "mean(Rating)",
    x = "Year Released",
    color = alt.condition(selection, alt.value("orange"), alt.value("lightgrey"))
).add_selection(selection).properties(height=250, width=250)

# Specify our detail chart
detail = base.mark_bar().encode(
    y = "Rating",
    x = "Title"
).transform_filter(selection).properties(height=250, width=250)

overview | detail
```

```python
# Implement selection
selection = alt.selection(type="multi", fields=["Game Type"])

alt.Chart(data).mark_circle().encode(
    x = alt.X("Year Released", scale=alt.Scale(domain=[1997,2024])),
    y = "Rating",
    color = alt.Color("Game Type", scale=alt.Scale(scheme='spectral')),
    size = "Plays",
    tooltip=["Title", "Rating", "Year Released"],
    opacity=alt.condition(selection, alt.value(1), alt.value(.2))
).add_selection(selection)
```
