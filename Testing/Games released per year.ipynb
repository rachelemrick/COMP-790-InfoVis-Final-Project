```python
import pandas as pd
import altair as alt

data = pd.read_csv("./Databases/compiled_db.csv")
data.head()
```

```python
df = pd.DataFrame(data)

genre_list = ['', 'Adventure', 'RPG', 'Indie', 'Turn Based Strategy', 'Platform', 'Simulator', 'Brawler', 'Puzzle', 'Shooter', 'Music', 'Strategy', 
              'Fighting', 'Visual Novel', 'Tactical', 'Racing', 'Arcade', 'Point-and-Click', 'Sport', 'Card & Board Game', 'Real Time Strategy']
genre_dropdown = alt.binding_select(options=genre_list, name="Select a genre to filter (empty selection to see all games): ")
genre_selector = alt.param(name='SelectorName', value='', bind=genre_dropdown)


# Group by 'Year Released' and 'Game Type' and count occurrences
df_grouped = df.groupby(['Year Released', 'Game Type']).size().reset_index(name='Games Released')

# Creating the line chart
chart = alt.Chart(df_grouped).mark_line(point=True).encode(
    x='Year Released:O',
    y='Games Released:Q',
    color='Game Type:N',
    tooltip=['Year Released', 'Game Type', 'Games Released']
).properties(
    width=600,
    height=400,
    title='Yearly Game Releases for Indie vs AAA games'
).add_params(genre_selector).interactive()

chart.display()
```

```python
import pandas as pd
import altair as alt

dropdown = alt.binding_select(options=["Indie", "AAA"], name="Select a type to filter: ")
selection = alt.selection_single(fields=["Game Type"], bind=dropdown, name="Type")

# Split genres into individual rows
df_split_genres = df.assign(Genre=df['Genres'].apply(lambda x: x.strip("[]").replace("'", "").split(", "))).explode('Genre')

# Group by 'Year Released', 'Genre', and 'Game Type', and count occurrences
df_grouped = df_split_genres.groupby(['Year Released', 'Genre', 'Game Type']).size().reset_index(name='Games Released')

# Filter data based on selection
filtered_data = alt.FieldEqualPredicate(field='Game Type', equal='Indie')

chart = alt.Chart(df_grouped).mark_line(point=True).encode(
    x='Year Released:O',
    y='Games Released:Q',
    color='Genre:N',
    tooltip=['Year Released', 'Genre', 'Games Released']
).transform_filter(
    selection
).properties(
    width=600,
    height=400,
    title='Yearly Game Releases for Indie vs AAA games by Genre'
).add_selection(
    selection
).interactive()

chart.display()
```
